```python
import seaborn as sns
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=tips) #依照禮拜幾打印bill 
```

```python
#加入分類變數, 用hue做屬性

sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips)
```

```python
tips = sns.load_dataset("tips")
ax = sns.boxplot(x="sex", y="total_bill", data=tips) #性別做相形圖
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="Set3")
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="sex", data=tips, linewidth=1.5) #x = day, y = total, hue = class 
```

```python

bill = tips["total_bill"]
ax = sns.distplot(bill)#total_bill as x asis
```

```python
#bill number as x asis value, tip as y asis value , draw the scatter plot 
sns.jointplot(x="total_bill", y="tip", data=tips)  
```

```python
#displot繪製單一變數的資料分配情況，jointplot繪製雙變數的分配情況，而pairplot則可以對一資料集繪製兩兩變數之間的分配關係
sns.pairplot(tips)
```
